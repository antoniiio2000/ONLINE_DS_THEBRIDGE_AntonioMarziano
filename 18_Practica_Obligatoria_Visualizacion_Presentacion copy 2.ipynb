### #1.2

Crea una función similar a la que se pide en el ejercicio 6 de la segunda tanda de ejercicios del workout. Es decir, una función para pintar diagramas de dispersion que tenga seis argumentos, pero ojo al cambio en el quinto:
1. Un dataframe
2. Una columna numérica del dataframe
3. Una columna numérica del dataframe diferente a la anterior.
4. Un columna categórica (Opcional, valor por defecto None)
5. Un valor para el tamaño de los puntos o una cadena, pero que puede ser una cadena de caracteres, en cuyo caso se debe interpretar como una de las columnas del dataframe cuyos valores deben controlar el tamaño
6. Un sexto argumento "escala", que permita manipular la escala del tamaño en el caso de que el quinto argumento sea una cadena. Este argumento multiplicará a los valores de la columna indicada en el parámetro anterior (así si quiero dividir por 10000, este sexto argumento tendrá que valer 1/10000)

La función debe pintar el diagrama de dispersión de la columna numérica 1, en el eje X, frente a la columna numérica 2, en el eje Y, usando un código de colores en función de la columna categórica y con el tamaño de los puntos según el argumento correspondiente. 

Si no se diera una columna categórica como argumento, el color no debe pasarse como parámetro a la función de Seaborn o Matplotlib que uses.

EXTRA: Añade un argumento que controle si la función pinta una leyenda para colores y tamaños o un colorbar y una leyenda para los tamaños. NO TE PREOCUPES POR EL FORMATO DE LA LEYENDA (lo veremos en la sesión en vivo)

```python

def scatter_plot(df, x_col, y_col, hue=None, size_col=None, scale=1, show_legend=True):
    """
    Esta función genera un diagrama de dispersión con los parámetros especificados.

    :param df: DataFrame que contiene los datos.
    :param x_col: Columna numérica para el eje X.
    :param y_col: Columna numérica para el eje Y.
    :param hue: Columna categórica para color, puede ser None.
    :param size_col: Columna numérica o tamaño fijo para los puntos.
    :param scale: Factor para ajustar la escala del tamaño.
    :param show_legend: Si True, muestra leyenda de colores y tamaños.
    """
    
    plt.figure(figsize=(10, 6))
    
    # Si se proporciona una columna categórica para el color
    if hue:
        scatter = sns.scatterplot(data=df, x=x_col, y=y_col, hue=hue, size=size_col, sizes=(20, 200), palette="viridis", scale=scale, legend=show_legend)
    else:
        scatter = sns.scatterplot(data=df, x=x_col, y=y_col, size=size_col, sizes=(20, 200), scale=scale, legend=show_legend)
    
    plt.title(f"Diagrama de dispersión entre {x_col} y {y_col}", fontsize=16)
    plt.xlabel(x_col, fontsize=12)
    plt.ylabel(y_col, fontsize=12)
    plt.show()

```

### #1.3

Pinta el diagrama de dispersión de longitud y latitud (longitud en las x, latitud en las y) de las ciudades, usando el tamaño de las ciudades para el tamaño de los puntos y los colores para mostrar la población. Previamente crea una columna nueva "log_poblacion" y asignale el log decimal de la poblacion (np.log10() y como argumento los valores de la columna "population_total"). Usa "log_poblacion" para el color.

Un ejemplo de como podría salir para que lo tengas de referencia: (olvidate de la leyenda, auque si quieres como extra puedes intentar ver cómo crearla así)

```python
# Crear columna log_poblacion
df_cities["log_poblacion"] = np.log10(df_cities["population_total"])

# Crear gráfico de dispersión
plt.figure(figsize=(10, 6))

# Usamos longitud en X, latitud en Y, tamaño basado en la población y color basado en log_poblacion
scatter = sns.scatterplot(data=df_cities, x="longd", y="latd", size="population_total", sizes=(20, 200), hue="log_poblacion", palette="coolwarm", legend=False, alpha=0.7)

plt.title("Diagrama de dispersión: Longitud vs Latitud", fontsize=16)
plt.xlabel("Longitud", fontsize=12)
plt.ylabel("Latitud", fontsize=12)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset Titanic
df_titanic = sns.load_dataset('titanic')

# Ejemplo de gráfico: tasa de supervivencia por clase
sns.barplot(x="class", y="survived", data=df_titanic)
plt.title("Tasa de supervivencia por clase")
plt.xlabel("Clase")
plt.ylabel("Supervivencia")
plt.show()

# Ejemplo de gráfico: tasa de supervivencia por sexo
sns.barplot(x="sex", y="survived", data=df_titanic)
plt.title("Tasa de supervivencia por sexo")
plt.xlabel("Sexo")
plt.ylabel("Supervivencia")
plt.show()

# Ejemplo de gráfico: tasa de supervivencia por puerto de embarque
sns.barplot(x="embarked", y="survived", data=df_titanic)
plt.title("Tasa de supervivencia por puerto de embarque")
plt.xlabel("Puerto de embarque")
plt.ylabel("Supervivencia")
plt.show()
```
